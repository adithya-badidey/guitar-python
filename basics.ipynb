```python
pentamajscale = [0, 2, 4, 7, 9]
pentamaj = NoteSet(pentamajscale, Note('G2'))
print(f"Notes: {pentamaj}")
print()
guitar.printFretboardDiagram(pentamaj)
mode = pentamaj
print("Chords:")
for i in range(len(mode.prog)):
  note = mode.getRootNoteIndex()
  for i in chords:
    chord = NoteSet(chords[i], Note(note))
    if chord in mode:
      print(f"{iton[note]} {i}: {chord}")
  mode = changeRootNote(mode, 1)
```
